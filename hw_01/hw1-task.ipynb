```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

#TODO
```

```python
def f1(s):
    if (s<=10000):
        return 'A'
    elif (10000<s<=100000):
        return 'B'
    elif (100000<s<=200000):
        return 'C'
    elif (200000<s<=400000):
        return 'D'
    elif (400000<s<=700000):
        return 'E'
    elif (s>700000):
        return 'F'
```

```python
df['new_col']=df['LIMIT_BAL'].map(f1)
```

```python
print('Решение задание (6):')
print('A:', df[df.new_col=='A'].shape[0], 'B:', df[df.new_col=='B'].shape[0], 'C:', df[df.new_col=='C'].shape[0], 'D:', df[df.new_col=='D'].shape[0], 'E:', df[df.new_col=='E'].shape[0], 'F:', df[df.new_col=='F'].shape[0])
```
